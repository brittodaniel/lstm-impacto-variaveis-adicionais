```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
import statistics
```

```python
# Modelo LSTM Univariável

#valores
dados = np.sort([
    2444.16,
    2380.58,
    2196.75,
    2534.27,
    2463.71,
    2289.70,
    2713.56,
    2450.10,
    2698.69,
    2500.92
])

# média e desvio padrão
media = statistics.mean(dados)
desvio = statistics.stdev(dados)

# histograma
plt.hist(dados)
plt.suptitle('Modelo LSTM Univariável')
plt.title('Média: %.2f' % media + ' - Desvio Padrão: %.2f' % desvio)
plt.savefig('histo-uni.png', bbox_inches='tight')
plt.show()
```
